```python
'''
Given a positive integer, return its corresponding column title as appear in an Excel sheet.

For example:

    1 -> A
    2 -> B
    3 -> C
    ...
    26 -> Z
    27 -> AA
    28 -> AB 
'''
```

```python
class Solution(object):
    def convertToTitle(self, n):
        """
        :type n: int
        :rtype: str
        """
        colum = []
        while n > 0:
            if n%26 == 0:
                colum.append(26)
                n = (n-1)//26
            else:
                r = n%26
                n = n//26
                colum.append(r)
        returnAns = ''
        for c in colum[::-1]:
            returnAns += chr(c+64)
        return returnAns
```

```python
class Solution2(object):
    def convertToTitle(self, num):
        result = ""
        while num > 0:
            if num % 26 == 0:
                result = "Z" + result
                num = int(num / 26) - 1
            else:
                result = chr((num % 26) + 64) + result
                num = int(num / 26)
        return result
```

```python
class Solution3(object):
    def convertToTitle(self, n):
        """
        :type n: int
        :rtype: str
        """
        result = []
        base = ord('A')
        while n:
            n, r = divmod(n - 1, 26)
            result.append(chr(base + r))
        return ''.join(result[::-1])
```
